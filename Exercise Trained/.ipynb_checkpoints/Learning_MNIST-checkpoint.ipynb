```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import plot_model
from tensorflow.keras.datasets import mnist
```

```python
# Descargar el conjunto de datos MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Visualizar algunas imágenes de entrenamiento
fig, axs = plt.subplots(2, 4, figsize=(8, 4))
for i, ax in enumerate(axs.flat):
    ax.imshow(x_train[i])
    ax.set_title(f"Label: {y_train[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Mostrar la forma de los tensores
print(x_train.shape)
print(y_train.shape)
```

```python
#Amoldar la forma del tensor para introducir al modelo
x_train = x_train.reshape(-1,28, 28, 1)
x_test = x_test.reshape(-1,28, 28, 1)

#Neuva forma
print(x_train.shape)
print(x_test.shape)
```

```python
# Definir la arquitectura del modelo

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3), padding="valid", activation="relu",name="1_CNN",input_shape=(28,28,1)),
    tf.keras.layers.MaxPooling2D((2, 2),name="Maxpooling_2x2"),
    tf.keras.layers.Conv2D(64, (3, 3), strides=(2,2), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation="relu"),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(10,activation="softmax")
])
model.summary()
```

```python
#Entrenamiento

# Compilar el modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

```python
#Salvar el modelo para usarlo posteriormente
model.save('modelo.h5')

#Recuperar el modelo
model_recuperado = tf.keras.models.load_model('modelo.h5')
```

Primero obtenemos las predicciones del modelo en el conjunto de prueba utilizando el método predict(). Luego, utilizamos la función argmax() de NumPy para obtener las etiquetas predichas por el modelo.

A continuación, creamos una figura con una cuadrícula de subplots utilizando la función subplots() de Matplotlib. Iteramos sobre los subplots y mostramos las imágenes de prueba correspondientes, junto con las etiquetas predichas y las etiquetas reales.

Finalmente, calculamos la precisión del modelo en el conjunto de prueba comparando las etiquetas predichas con las etiquetas reales y calculando el porcentaje de aciertos.

```python
# Obtener las predicciones del modelo en el conjunto de prueba
predictions = model.predict(x_test)
predicted_labels = np.argmax(predictions, axis=1)

# Mostrar imágenes de prueba con las etiquetas predichas y las etiquetas reales
fig, axes = plt.subplots(5, 5, figsize=(7, 7))
axes = axes.flatten()

for i, ax in enumerate(axes):
    ax.imshow(x_test[i], cmap='gray')
    ax.axis('off')
    ax.set_title(f'Pred:{predicted_labels[i]}, Real:{y_test[i]}')

plt.tight_layout()
plt.show()

# Calcular la precisión del modelo en el conjunto de prueba
accuracy = np.sum(predicted_labels == y_test) / len(y_test)
print('Precisión en el conjunto de prueba: ', round(accuracy*100,2),"%")
```

### Creación de ruido para mejorar entrenamiento
Una técnica tambien usada para mejorar la eficacia y fortaleza de los modelos es entrenarlos con las imagenes con cierta distorisión (ruido), esto hace que los niveles de validación sean más exigentes y por lo tanto mejora la eficacia de nuestro modelo frente a imagenes con una buen visualización

```python
noise_factor = 0.6
Data_noisy = x_train + noise_factor * np.random.normal(size=x_train.shape) 

#Asegurarse de que ese ruido queda dentro del rango [0,1]
Data_noisy = np.clip(Data_noisy, 0.0, 1.0)
```

```python
#Visualización del ruido
n = 5
plt.figure(figsize=(10, 4.5))
for i in range(n):
    # plot original image
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    if i == n/2:
        ax.set_title('Original Images')

    # plot noisy image 
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(Data_noisy[i], cmap="gray")
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    if i == n/2:
        ax.set_title('Noisy Input')
```
